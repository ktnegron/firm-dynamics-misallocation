# Comparison: Base Model vs. Credit Extension

**When does targeting help? It depends entirely on whether there's a real friction.**

This notebook compares the two models side-by-side to highlight the central insight:

- **Base model** (wedge/tax-subsidy): Targeting misallocates an otherwise-efficient equilibrium → always bad
- **Credit model** (collateral constraint): Targeting relaxes a genuine friction → can be good

## 1. Load Results from Both Models

```python
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Load base model results
try:
    base_table = pd.read_csv('results/base/results_table.csv')
    print("✓ Loaded base model results from results/base/results_table.csv")
except FileNotFoundError:
    print("⚠ Base model results not found. Run 01_base_model.ipynb first.")
    base_table = None

# Load credit model results
try:
    credit_table = pd.read_csv('results/credit/results_table.csv')
    print("✓ Loaded credit model results from results/credit/results_table.csv")
except FileNotFoundError:
    print("⚠ Credit model results not found. Run 02_credit_model.ipynb first.")
    credit_table = None

if base_table is None or credit_table is None:
    print("\n⚠ Please run both 01_base_model.ipynb and 02_credit_model.ipynb first.")
```

## 2. Base Model: Output Loss Summary

```python
if base_table is not None:
    print("BASE MODEL: Aggregate Output Relative to Efficient Benchmark")
    print("="*70)
    print(base_table[['Regime', 'Output (% of efficient)', 'Output Loss (%)']].to_string(index=False))
    print()
    print("Key observation:")
    print("  • Even UNCORRELATED distortions cost 11% output (Restuccia-Rogerson result)")
    print("  • MISTARGETING costs 31% output — much worse")
    print("  • Why? Subsidized weak firms stay alive, crowding out strong firms.")
```

## 3. Credit Model: Policy Effectiveness

```python
if credit_table is not None:
    print("CREDIT MODEL: Output Relative to Frictionless Benchmark")
    print("="*70)
    print(credit_table[['Regime', 'Output (% of frictionless)', '% of friction gap closed']].to_string(index=False))
    print()
    print("Key observation:")
    print("  • TARGETED policy closes ~50% of the friction gap")
    print("  • MISTARGETED policy closes 0% (wasted, but not harmful)")
    print("  • Why the difference from base model? Relaxing credit is optional;")
    print("    weak firms simply don't borrow. But in the base model, taxes FORCE distortion.")
```

## 4. Side-by-Side Comparison: The Role of Frictions

```python
if base_table is not None and credit_table is not None:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Base model output levels (efficient = 100)
    base_regimes = ['Efficient', 'Untargeted', 'Mistargeted']
    base_outputs = base_table['Output (% of efficient)'].values
    colors_base = ['#2E75B6', '#E69F00', '#C0392B']
    
    ax1.bar(base_regimes, base_outputs, color=colors_base, width=0.6)
    ax1.axhline(100, color='grey', linestyle='--', linewidth=0.8)
    ax1.set_ylabel('Output (efficient = 100)', fontsize=11)
    ax1.set_ylim(0, 110)
    ax1.set_title('BASE MODEL\n(wedge/tax-subsidy, no real friction)', fontsize=11, fontweight='bold')
    ax1.grid(axis='y', alpha=0.3)
    
    for i, (regime, val) in enumerate(zip(base_regimes, base_outputs)):
        ax1.text(i, val + 3, f'{val:.1f}', ha='center', fontsize=10)
    
    # Credit model output levels (frictionless = 100)
    credit_regimes = ['Frictionless', 'Constrained', 'Targeted', 'Mistargeted']
    credit_outputs = credit_table['Output (% of frictionless)'].values
    colors_credit = ['#2E75B6', '#8C8C8C', '#2E9E5B', '#C0392B']
    
    ax2.bar(credit_regimes, credit_outputs, color=colors_credit, width=0.6)
    ax2.axhline(100, color='grey', linestyle='--', linewidth=0.8)
    ax2.set_ylabel('Output (frictionless = 100)', fontsize=11)
    ax2.set_ylim(0, 110)
    ax2.set_title('CREDIT MODEL\n(collateral constraint: real friction)', fontsize=11, fontweight='bold')
    ax2.grid(axis='y', alpha=0.3)
    
    for i, (regime, val) in enumerate(zip(credit_regimes, credit_outputs)):
        ax2.text(i, val + 3, f'{val:.1f}', ha='center', fontsize=10)
    
    plt.xticks(fontsize=9)
    fig.suptitle('The Role of Real Frictions: Targeting Effectiveness Comparison', 
                 fontsize=13, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.savefig('../results/comparison.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("[OK] Saved: ../results/comparison.png")
```

## 5. The Policy Implication

```python
policy_lesson = """
╔════════════════════════════════════════════════════════════════════════════╗
║                          THE CENTRAL INSIGHT                              ║
╚════════════════════════════════════════════════════════════════════════════╝

When does "pick the winners" work as development policy?

  ❌ BASE MODEL (no real friction):
     • The efficient, undistorted equilibrium is already first-best.
     • Any selective subsidy/tax creates misallocation IN A NEW DIRECTION.
     • Targeting productive firms → forces them past their efficient size.
     • Result: WORSE than doing nothing. (~11% loss from untargeted,
       ~31% loss from mistargeted)

  ✓ CREDIT MODEL (real, pre-existing friction):
     • Some talented firms are stuck below efficient scale due to credit constraint.
     • Relaxing the constraint FOR THOSE FIRMS closes a genuine inefficiency.
     • Mistargeting doesn't help (weak firms don't want the credit).
     • But it also doesn't HURT — it's just wasted.
     • Result: Targeted policy WORKS. (~50% of friction gap closed)

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

DEVELOPMENT IMPLICATION:

  Before applying selective credit/tax policies, ask:
    1. Is there a REAL market failure (financing constraint, externality, etc.)?
    2. Does the policy target the right margin (where the friction binds)?
    3. Is the targeting ACCURATE (can you really identify talent vs. weakness)?

  If (1) is YES and (2)+(3) are YES → Targeting can help significantly.
  If (1) is NO or (2)+(3) are NO → Selective policy likely hurts aggregate output.

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

LITERATURE:
  • Hopenhayn (1992): Firm-dynamics equilibrium model
  • Restuccia & Rogerson (2008): Policy distortions and productivity
  • Buera, Kaboski & Shin (2011): Credit constraints and development
  • Hsieh & Klenow (2009): Misallocation and productivity
"""

print(policy_lesson)
```
